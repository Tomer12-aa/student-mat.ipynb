```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('student-mat.csv', sep=';')
df.head()
```

```python
#Explore & Clean Data
#Check Dataset Shape
df.shape
```

```python
#Check Data Types
df.dtypes
```

```python
#Check Missing Values
df.isnull().sum()
```

```python
#Remove Duplicate Rows
df = df.drop_duplicates()
```

```python
# check again shape

df.shape
```

```python
# Analysis Questions
#Average Final Grade (G3)
average_grade = df["G3"].mean()
print("Average Final Grade:", average_grade)
```

```python
#How Many Students Scored Above 15
above_15 = df[df["G3"] > 15].shape[0]
print("Students scoring above 15:", above_15)
```

```python
#Correlation Between Study Time and Performance
correlation = df["studytime"].corr(df["G3"])
print("Correlation between Study Time and Final Grade:", correlation)
```

```python
#Which Gender Performs Better on Average
df.groupby("sex")["G3"].mean()
```

```python
#Visualizations
#Histogram of Grades
plt.hist(df["G3"], bins=10)
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
```

```python
#Scatterplot: Study Time vs Grades
plt.scatter(df["studytime"], df["G3"])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

```python
#Bar Chart: Male vs Female Average Score
avg_scores = df.groupby("sex")["G3"].mean()

avg_scores.plot(kind="bar")
plt.title("Average Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.show()
```
